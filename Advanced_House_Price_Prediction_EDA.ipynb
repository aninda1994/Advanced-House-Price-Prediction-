# Missing value

```
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
  print(feature, np.round(dataset[feature].isnull().mean(),4),  "%missing values")
```

```
for feature in features_with_na:
  data = dataset.copy()
  data[feature] = np.where(data[feature].isnull(), 1,0)
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()


```

Here we clearly see that the missing value has an impact on our dependent variable so we have to deal the missing value in feature engineering section.


Id is unique value so we have to drop it in feature engineering section.

# Numerical variable

```
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print("Number of numerical variable - ", len(numerical_feature))
dataset[numerical_feature].head()
```

```
year_feature = [feature for feature in numerical_feature if 'Year' in feature or 'Yr' in feature]
print(year_feature)
```

```
for feature in year_feature:
  print(feature, dataset[feature].unique())
```

```
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Sale Price')
plt.title('House price vs Year Sold')
```

```
for feature in year_feature:
  if feature != 'YrSold':
    data = dataset.copy()
    data[feature] = data['YrSold']-data[feature]
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel("Sale Price")
    plt.show()
```

```
discrete_feature = [feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]

print('Discrete variables count: {}'.format(len(discrete_feature)))
```
